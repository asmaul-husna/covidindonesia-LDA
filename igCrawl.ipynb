```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import os
import  wget
import pandas as pd
import numpy as np
import time
import seaborn as sns
from matplotlib import pyplot as plt

driver = webdriver.Chrome("F:/chromedriver.exe")#directory chromedriver, mau dijalanin sama script
driver.get("https://instagram.com")

#csv
collected_data = {'username':[],'text':[],'date':[]}
collected_data = pd.DataFrame(collected_data)
collected_data

print (len(collected_data))

#ini mulai masuk javascript
#driver.execute_script ("var scroll_top = 0; var scroll_bot = 1000; for (i=0, i <10, i++){window.scrollTo(scroll_top,scroll_bot); scroll_top = scroll_bot; scroll_bot+=1000;}")
#problem = pasti ada loop limit
counter = 0
for j in range(0,400):
    selected_post = driver.find_elements_by_xpath("//*[@class='_9AhH0']")
    #div clas _9AhH0
    #driver.find_elements_by_tag_name('h2')<-- ini gk mungkin, soalnya hidden
    len(selected_post)
    for i in range(counter,len(selected_post)) :
        selected_post[i].click()
        #nama uploader
        #class = sqdOP yWX7d     _8A5w5   ZIAjV 
        post_name = WebDriverWait(driver,60).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "a[class='sqdOP yWX7d     _8A5w5   ZIAjV ']")))
        post_name = post_name.text
        print (post_name)
        #delete username element in post
        driver.execute_script("var element = document.getElementsByClassName('_6lAjh ');while (element.length > 0) element[0].remove();")
        #text post
        #div class =C4VMK
        post_text = WebDriverWait(driver,60).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "div[class='C4VMK']")))
        #post_text = post_text.find_elements_by_xpath("//span")
        post_text = post_text.text
        print (post_text)
        #waktu posting
        #class = FH9sR Nzb55
        post_date = WebDriverWait(driver,60).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "time[class='FH9sR Nzb55']")))
        post_date = post_date.get_attribute('datetime')
        print(post_date)
        #insert to collected data
        collected_data.loc[len(collected_data)] = [post_name, post_text, post_date]
        #exit post
        
        #class =wpO6b  
        some_button = driver.find_elements_by_xpath("//*[@class='wpO6b  ']")
        exit_post = some_button[-1].click()#1 is frozen var-do not edit, 0 is report button
        print (len (some_button))
        time.sleep(1)
    counter = len(selected_post)
    print ("iterasi ",j, ", counter = ", counter)
    driver.execute_script ("window.scrollTo(0,1000000)")
    time.sleep(5)
    
print(collected_data.duplicated().sum())
print(collected_data.count())
print(collected_data)

#data cleaning, hapus enter
copied_data = collected_data.copy()

for a in range(0,len(copied_data)):
    copied_data.iloc[a]['text']=copied_data.iloc[a]['text'].replace("\n", "")
    #print(copied_data.iloc[a]['text'])

print(copied_data.iloc[0]['text'])
print(copied_data.iloc[741]['date'])
copied_data.to_csv("nameyourfile.csv")
```
